```python
from bs4 import BeautifulSoup
import json

def htmlScraper(filename):
    with open(filename) as f:
        content = f.read()
        soup = BeautifulSoup(content, 'html.parser')
        reviews = soup.find_all(class_="review-container")
    # print(reviews[0])
    res = []
    for review in reviews:
        date = review.find('span', attrs={'class': 'review-date'}).text
        name = review.find('span', attrs={'class': 'display-name-link'}).text
        title = review.find('a', attrs={'class': 'title'}).text
        comment = review.find('div', attrs={'class': 'text'}).text
        rating = str(review.span.span)
        if rating!='None':
            if rating[7]<='9':
                rating = rating[6:8]
            else:
                rating = rating[6:7]
        data = {'Date': str(date), 'Name': str(name), 'Title': str(title), 'Comment': str(comment), 'Rating': str(rating)}
        res.append(data)
    title = soup.find_all(itemprop='name')[0].a.getText()
    print(title)
    with open('Dataset/' + title + '.json', 'w') as f:
        json.dump(res, f)
```

```python
htmlScraper('bots/selenium/121804.html')
```

```python
from pathlib import Path
p = Path('HTML-Files/')
for filePath in p.iterdir():
    if not filePath.is_dir():
        try:
            htmlScraper(str(filePath))
        except:
            print('Exception occured: {}'.format(str(filePath)))
    
```

```python
from pathlib import Path
import json
jsonPath = Path('Dataset/')

masterDict = dict()
for filePath in jsonPath.iterdir():
    if not filePath.is_dir():
        with open(filePath,'r') as f:
            jsonData = json.load(f)
        sumRating = 0
        filtJsonData = []
        try:
            for comment in jsonData:
                if comment['Rating'].isdecimal():
                    comment['Rating'] = int(comment['Rating'])
                    filtJsonData.append(comment)
                    sumRating += comment['Rating']
            n = len(filtJsonData)
            rating = sumRating / n
            masterDict[filePath.name[:-5]] = {
                'Comments' : filtJsonData,
                'Size' : n,
                'Rating' : rating
            }
            print("{:<20} {:<5} {:<5.3}".format(filePath.name,n,rating))
        except:
            print("Exception occured {}".format(filePath.name))
            continue
        
        
with open('dataset.json','w') as dset:
    json.dump(masterDict,dset)
            
```
